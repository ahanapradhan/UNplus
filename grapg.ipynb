```python

# importing package
import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = np.arange(5)
y1 = [0.7909164428710938,  0.8257734775543213 ,0.8809921741485596 , 0.8189539909362793, 0.847132682800293] # correlated sampling
y2 = [(1.4464843273162842 + 0.7909164428710938),  (4.807533025741577 + 0.8257734775543213), (4.089139461517334 + 0.8809921741485596),(6.429215431213379+0.8189539909362793), (2.960028648376465 + 0.847132682800293)] #cop[y based samp]

y3 = [12.611318111419678,  9.540351629257202 ,30.049304723739624 , 22.069141149520874, 13.986800909042358] #only copy based sampling
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x, y1, width, color='red')
plt.bar(x, y2, width,bottom=y1, color='green')
plt.bar(x+0.2, y3, width, color='cyan')
plt.xticks(x, ['Q1', 'Q2', 'Q3', 'Q5', 'Q17'])
plt.xlabel("Queries")
plt.ylabel("Run Time(Seconds)")
plt.legend(["Correlated S(CS)", "Copy-Based M. after CS", "Only Copy Based M."],  loc="upper left")
plt.show()
```
